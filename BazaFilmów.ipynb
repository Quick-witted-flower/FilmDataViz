1. Wyświetlam listę 10 najwyzej ocenianych filmów (vote_average)których liczba głosów (vote_count) jest większa od 3. kwartyla rozkładu liczby głosów.


```python
import pandas as pd
import matplotlib.pyplot as plt

genres_df = pd.read_csv('tmdb_genres.csv')
movies_df = pd.read_csv('tmdb_movies.csv')

display(genres_df.head())
display(movies_df.head())

third_quartile_vote_count = movies_df['vote_count'].quantile(0.75)
print("3. kwartyl liczby głosów:", third_quartile_vote_count)

filtered_movies = movies_df[movies_df['vote_count'] > third_quartile_vote_count]
top_10_movies = filtered_movies.sort_values(by='vote_average', ascending=False).head(10)

display(top_10_movies)
```

Odp1.Wykres pokazuje Top 10najwyżej ocenianych filmów(vote_count > 737)

```python
plt.plot(top_10_movies['title'], top_10_movies['vote_average'], marker='o')
plt.title('Top 10 najwyżej ocenianych filmów (vote_count > 737)')
plt.xticks(rotation=45, ha='right')
plt.xlabel('Tytuł filmu')
plt.ylabel('Średnia ocena')
plt.show()
```

2a.Grupuje tabelę w taki sposób, aby otrzymać średni przychód (revenue) oraz średni budżet (budget) w danym roku dla filmów opublikowanych od 2010 (włącznie) do 2016 roku (włącznie) i następnie ...(..)

```python
movies_filtered_df = movies_df[(movies_df['release_date'] >= '2010-01-01') & (movies_df['release_date'] <= '2016-12-31')].copy()

movies_filtered_df['release_date'] = pd.to_datetime(movies_filtered_df['release_date'])
movies_filtered_df['release_year'] = movies_filtered_df['release_date'].dt.year

movies_grouped_by_year = movies_filtered_df.groupby('release_year').agg({'revenue': 'mean', 'budget': 'mean'}).reset_index()

print(movies_grouped_by_year)
```

2b.(...)na podstawie tabeli tworzę wykres, w którym średnie przychody są wykresem kolumnowym, a średni budżet wykresem liniowym na tych samych osiach. Sformatuje odpowiednio oś X oraz oś Y. Dodaje tytuł wykresu, oraz legendę, która znajduje się w prawym górnym rogu płótna, lecz poza obszarem osi.

```python
years = movies_grouped_by_year['release_year']
average_revenue = movies_grouped_by_year['revenue']
average_budget = movies_grouped_by_year['budget']

fig, ax = plt.subplots(figsize=(10, 6))

ax.bar(years, average_revenue, label='revenues', alpha=0.7)

ax.plot(years, average_budget, color='red', label='budget', linewidth=2)

ax.set_title('Średni przychód i budżet filmu w latach 2010-2016 roku')
ax.set_xlabel('Rok')
ax.set_ylabel('Średnia wartość (USD)')

ax.get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x*1e-6:.1f}M'))

ax.legend(loc='upper left', bbox_to_anchor=(1, 1))

plt.tight_layout()
plt.show()
```

4.Opowiadam na pytania: jaki gatunek filmu z bazy pojawia się w niej najczęściej? Ile filmów tego gatunku znajduje się w bazie?

```python
genres_df = genres_df.rename(columns={'Unnamed: 0': 'genre_id', 'genres': 'genre_name'})
movies_with_genres_df = pd.merge(movies_df, genres_df, on='genre_id', how='left')

genre_counts = movies_with_genres_df['genre_name'].value_counts()


plt.figure(figsize=(12, 8))
genre_counts.plot(kind='bar', color='skyblue')
plt.title('Liczba filmów w każdym gatunku')
plt.xlabel('Gatunek')
plt.ylabel('Liczba filmów')
plt.xticks(rotation=45)
plt.show()
```

5.Odpowiadam na pytanie : Filmy, którego gatunku trwają średnio najdłużej (runtime)?

```python
genres_df = genres_df.rename(columns={'Unnamed: 0': 'genre_id', 'genres': 'genre_name'})
movies_with_genres_df = pd.merge(movies_df, genres_df, on='genre_id', how='left')

average_runtime_by_genre = movies_with_genres_df.groupby('genre_name')['runtime'].mean().reset_index()
longest_average_runtime_genre = average_runtime_by_genre.sort_values(by='runtime', ascending=False).iloc[0]

print("Gatunek filmów o najdłuższym średnim czasie trwania to:", longest_average_runtime_genre['genre_name'])
print("Średni czas trwania (w minutach):", round(longest_average_runtime_genre['runtime'], 2))

average_runtime_by_genre = average_runtime_by_genre.sort_values(by='runtime', ascending=False)

plt.figure(figsize=(12, 8))
plt.bar(average_runtime_by_genre['genre_name'], average_runtime_by_genre['runtime'], color='skyblue')
plt.title('Średni czas trwania filmów w różnych gatunkach')
plt.xlabel('Gatunek')
plt.ylabel('Średni czas trwania (minuty)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

6. Drukuje histogram czasu trwania filmów z gatunku, który cechuje się największym średnim czasem trwania.


```python
genres_df = genres_df.rename(columns={'Unnamed: 0': 'genre_id', 'genres': 'genre_name'})
movies_with_genres_df = pd.merge(movies_df, genres_df, on='genre_id', how='left')

average_runtime_by_genre = movies_with_genres_df.groupby('genre_name')['runtime'].mean().reset_index()

longest_average_runtime_genre = average_runtime_by_genre.sort_values(by='runtime', ascending=False).iloc[0]
longest_genre_name = longest_average_runtime_genre['genre_name']

movies_in_longest_genre = movies_with_genres_df[movies_with_genres_df['genre_name'] == longest_genre_name]

plt.figure(figsize=(10, 6))
plt.hist(movies_in_longest_genre['runtime'].dropna(), bins=15, color='skyblue', edgecolor='black')
plt.title(f'Histogram czasu trwania filmów w gatunku {longest_genre_name}')
plt.xlabel('Czas trwania (minuty)')
plt.ylabel('Liczba filmów')
plt.show()
```
